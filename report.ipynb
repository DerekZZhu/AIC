# Report + Graphs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_loss = [1.6196, 1.3967, 1.1905]
valid_loss = [1.4996, 1.39999, 1.3911]
```

```python
sns.lineplot(x=range(1,len(train_loss)+1), y=train_loss, markers="O")
sns.lineplot(x=range(1,len(valid_loss)+1), y=valid_loss)

plt.title("Training Loss for roBERTa")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```
